```python
#import city and ride data as data frames

city_csv = "city_data.csv"
city_data = pd.read_csv(city_csv)


ride_csv = "ride_data.csv"
ride_data = pd.read_csv(ride_csv)

ride_data


```

```python
#find total fares per city

ride_groups = ride_data.groupby('city')

total_fares_per_city = ride_groups["fare"].sum()
total_fares_per_city.head()
```

```python
#find ride count per city

ride_count_per_city = ride_groups["city"].count()
ride_count_per_city.head()
```
